### Merging DF ==> inner,outer,left,right
1. left: use only keys from left frame
2. right: use only keys from right frame
3. outer: use union of keys from both frames
4. inner: use intersection of keys from both frames
5. cross: creates the cartesian product from both frames

Example DF:
```
df1 = pd.DataFrame({"branch":['hyd','blr','vizag','pune'],
                   "total":[90,96,98,92]})
df2 = pd.DataFrame({"branch":['blr','hyd','vizag','chennai'],
```

### Merging

````python
df1 = pd.DataFrame({"branch":['hyd','blr','vizag','pune'],
                   "total":[90,96,98,92]})
df2 = pd.DataFrame({"branch":['blr','hyd','vizag','chennai'],
                   "distance_from_city":[40,60,12,20]})
````

````python
df1
````

````python
df2
````

````python
#inner ==> intersection of 2 DF
pd.merge(df1,df2)
````

````python
#outer ==> union of 2DF
pd.merge(df1,df2,how = "outer")
````

````python
#left
pd.merge(df1,df2,how = "left")
````

````python
pd.merge(df1,df2,how = "right")
````

````python
pd.merge(df1,df2,how = "cross")
````

### groupby

````python
data = {"name":['a','b','c','d','e','f'],
       "branch":['hyd','vizag','blr','hyd','vizag','blr'],
       "marks":[90,91,92,93,94,95]}
df = pd.DataFrame(data)
df
````

````python
for i,j in df.groupby("branch"):
    print(i)
    print(j)
````

````python
g = df.groupby("branch")
g
````

````python
g.get_group('blr')['marks'].sum()
````

````python
g.get_group('blr')['marks'].mean()
````

````python
g.sum()
````
