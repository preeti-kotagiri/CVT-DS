```python
sex=data.pop('Sex')
```

```python
data['male']=(sex==1)*1.0
data['female']=(sex==0)*1.0
```

```python
ExAng=data.pop('ExAng')
```

```python
data['a']=(ExAng==0)*1
data['b']=(ExAng==2)*1
data['c']=(ExAng==3)*1
data['d']=(ExAng==4)*1
```

```python
fbs=data.pop("Fbs")
```

```python
data['x']=(fbs==0)*1
data['y']=(fbs==1)*1
```

```python
rest=data.pop('RestECG')
```

```python
data['p']=(rest==0)*1
data['q']=(rest==1)*1
data['r']=(rest==2)*1
```

```python
slope=data.pop("Slope")
```

```python
data['f']=(slope==1)*1
data['g']=(slope==0)*1
data['h']=(slope==2)*1
```

```python
ca=data.pop("Ca")
```

```python
data['k']=(ca==0)*1
data['l']=(ca==1)*1
data['m']=(ca==2)*1
data['n']=(ca==3)*1
```

```python
thal=data.pop("Thal")
```

```python
data['o']=(thal==0)*1
data['t']=(thal==1)*1
data['u']=(thal==2)*1
data['v']=(thal==3)*1
```

```python
cp=data.pop('ChestPain')
```

```python
data['typical']=(cp=='typical')*1
data['asymptomatic']=(cp=='asymptomatic')*1
data['nonanginal']=(cp=='nonanginal')*1
data['nontypical']=(cp=='nontypical')*1
```

```python
y=data.AHD
x=data.drop("AHD",axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
import sklearn
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
from sklearn import ensemble
model=ensemble.AdaBoostClassifier(base_estimator=lr,n_estimators=100)
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_predict=model.predict(x_test)
print(y_predict)
```

```python
accuracy= accuracy_score(y_predict,y_test)
print(accuracy)
```

```python
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(y_predict,y_test)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
print(cm)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
print(np.mean(cross_val_score(model,x,y,cv=20,scoring="accuracy")))
```
