````python
import matplotlib.pyplot as plt
#%matplotlib inline
````

````python
import numpy as np
````

````python
#Lineplot ==> 2 var
x = np.arange(1,11)#1-10
y = np.arange(1,11)#1-10
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("plot shows relation b/w x and y")
plt.plot(x,y,color = "green",linewidth = 5)
plt.show()
````

````python
days = list(range(1,8))
temp = [34,35,36,37,36,35,34]
plt.plot(days,temp,color="k",linestyle="-.")
plt.show()
````
