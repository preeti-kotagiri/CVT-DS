```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
sub=['C', 'C++', 'Java', 'Python', 'PHP'] 
students=[23,17,35,29,12]
preference = [90,85,90,95,80]
extra = [10,20,30,40,50]
```

```python
xpos = np.arange(len(sub))
xpos
```

```python
plt.bar(xpos-0.2,students,width=0.2,label="students")
plt.bar(xpos+0.2,preference,width=0.2,label="preference")
plt.bar(xpos,extra,width=0.2,label='extra')
plt.xlabel("subjects")
plt.ylabel("students")
plt.title('sss')
plt.legend(loc='upper right')
plt.show()
```

### sub plot:

```python
fig,a = plt.subplots(2,2)
x = np.arange(1,5)
a[0][0].plot(x,2*x)
a[0][0].set_title("2 multiplied value")
a[0][1].plot(x,3*x)
a[0][1].set_title("3 multiplied value")
a[1][0].plot(x,4*x)
a[1][0].set_title("4 multiplied value")
a[1][1].plot(x,5*x)
a[1][1].set_title("5 multiplied value")
plt.show()
```

###  seaborn

```python
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
#scatterplot
sns.scatterplot(x ='sepal_length',y='sepal_width',data = iris)
```

```python
#pairplot
sns.pairplot(iris)
```

```python
#Linear regression plot
sns.lmplot(x ='sepal_length',y='sepal_width',data = iris)
```

```python
#box plot
sns.boxplot(x="species",y="sepal_length",data=iris)
```
