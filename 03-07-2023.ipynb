```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
email = pd.read_csv(r"C:\Users\hp\Desktop\CVT\Datasets-master\spam.csv",encoding='ISO-8859-1')
email.head()
```

```python
email.drop(["Unnamed: 2","Unnamed: 3","Unnamed: 4"],axis = 1,inplace = True)
```

```python
email.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(email.v2,email.v1,test_size=0.2,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
email.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
word_count_vector = count_vect.fit(X_train)
```

```python
len(word_count_vector.get_feature_names())
```

```python
doc_array = word_count_vector.transform(X_train)
doc_array
```

```python
doc_array = doc_array.toarray()
```

```python
feature_matrix = pd.DataFrame(doc_array,columns = word_count_vector.get_feature_names())
feature_matrix
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
model = BernoulliNB()
```

```python
model.fit(feature_matrix,y_train)
```

```python
y_train_pred = model.predict(feature_matrix)
```

```python
y_train_pred
```

```python
count_vect.fit(X_test).get_feature_names()
```

```python
feature_matrix1 = pd.DataFrame(count_vect.fit_transform(X_test).toarray(),columns = count_vect.fit(X_test).get_feature_names())
```

```python
feature_matrix1
```

```python
y_test_pred = model.predict(feature_matrix1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
X_train_transformed = tfidf.fit_transform(X_train)
X_train_transformed
```

```python
tfidf.get_feature_names()
```

```python
tfidf.idf_
```

```python
from sklearn.naive_bayes import BernoulliNB 
```

```python
model_NB = BernoulliNB()
```

```python
model_NB.fit(X_train_transformed,y_train)
```

```python
X_test_transformed = tfidf.transform(X_test)
```

```python
y_test_pred = model_NB.predict(X_test_transformed)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_test_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_test_pred))
```
