### Encoding using OneHotEncoder

```python
#import the OneHotEncoder class ==> sklearn.preprocessing
#create an object
#ColumnTransformer()==> we will pass OneHotEncoder() 
#two mandatory arg ==> transformers,remainder
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
col_trans = ColumnTransformer(transformers = [('encoder',OneHotEncoder(),[0])],remainder='passthrough')
```

```python
data = price.copy()
data
```

```python
data = col_trans.fit_transform(data)
data
```
